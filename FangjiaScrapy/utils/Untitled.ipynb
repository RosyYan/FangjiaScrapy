```python
from pyecharts import Map, Geo
```

```python
data = [('天河',30016), ('番禺',21000), ('越秀',29888), ('海珠',25676),( '荔湾',27331), ('白云',25677), ('南沙',21000),( '增城',19311), ('花都',24933), ('从化',19222)]
attr, value = geo.cast(data)
geo = Geo('广州各区二手房均价热力图','data from lianjia', title_color="#fff", title_pos = "center",width=1200,height=600,background_color="#404a59")
# map = Map("广州地图","广州", width=1200,height=600)
# map.add("广州", qu, values, visual_range=[1,10],maptype="广州",is_visualmap=True,visual_text_color='#000')

geo.add("均价热力图",attr, value, visual_range=[0,35000], type='heatmap',visul_text_color="#fff",symbol_size=15,is_visualmap=True, is_roam=False)
geo.show_config()
geo.render(path='./gz.html')
```
